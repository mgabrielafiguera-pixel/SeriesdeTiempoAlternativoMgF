```python
# Your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('https://breathecode.herokuapp.com/asset/internal-link?id=2546&path=sales.csv')
df
```

```python
df['date'] = pd.to_datetime(df['date'])
df = df.set_index('date').sort_index()
```

```python
print(df.index[1] - df.index[0])
print(pd.infer_freq(df.index))
```

```python
ts_sales = df['sales']
```

```python
# Grafica los datos originales y la tendencia suavizada
ts_sales.plot(label='Original')
ts_sales.rolling(window=12).mean().plot(label='Tendencia (Media Móvil)')
plt.legend()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=ts_sales)
plt.title("Evolución de Ventas Mensuales")
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Ejecutamos el test sobre la columna de ventas
result = adfuller(ts_sales)

print(f'Estadístico ADF: {result[0]}')
print(f'p-value: {result[1]}')
```
